```python
# ==================================================
# STEP 2: MISSING DATA ANALYSIS
# ==================================================

print("\n" + "=" * 70)
print("❓ MISSING DATA DETAILED REPORT")
print("=" * 70)

# Count missing values in each column
missing_count = df.isnull().sum()  # .isnull() checks if each cell is empty
                                    # .sum() counts True values (missing cells)

# Calculate percentage missing
missing_percent = (missing_count / len(df)) * 100  # Divide by total rows, multiply by 100

# Create a neat table
missing_df = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': missing_count.values,
    'Missing_Percent': missing_percent.values.round(2)  # Round to 2 decimals
})

# Sort by most missing first
missing_df = missing_df.sort_values('Missing_Count', ascending=False)

# Only show columns that have missing data
columns_with_missing = missing_df[missing_df['Missing_Count'] > 0]

if len(columns_with_missing) > 0:
    print("\n🔴 Columns with missing data:")
    print(columns_with_missing.to_string(index=False))

    print(f"\n📊 Summary:")
    print(f"   - Columns with missing data: {len(columns_with_missing)} out of {len(df.columns)}")
    print(f"   - Total missing cells: {missing_count.sum()}")
else:
    print("\n✅ NO MISSING DATA! Your dataset is complete!")

# Visualize missing data
print("\n📈 Generating missing data visualization...")

plt.figure(figsize=(12, 6))
columns_with_missing_sorted = columns_with_missing.sort_values('Missing_Count')
plt.barh(columns_with_missing_sorted['Column'],
         columns_with_missing_sorted['Missing_Count'],
         color='coral', edgecolor='black')
plt.xlabel('Number of Missing Values', fontsize=12, fontweight='bold')
plt.ylabel('Column Name', fontsize=12, fontweight='bold')
plt.title('Missing Data by Column', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ==================================================
# STEP 3: DATA CLEANING
# ==================================================

print("\n" + "=" * 70)
print("🧹 DATA CLEANING - HANDLING MISSING VALUES")
print("=" * 70)

# Create a copy to preserve original data
df_clean = df.copy()  # .copy() makes a duplicate so original df stays unchanged

print("\n🔹 Cleaning Strategy:")

# Strategy 1: Fill missing names with 'Unknown'
if 'Suspect Name' in df_clean.columns:
    missing_names = df_clean['Suspect Name'].isnull().sum()
    df_clean['Suspect Name'].fillna('Unknown', inplace=True)
    # fillna() = fill null/empty values
    # inplace=True = change the column directly (don't create new one)
    print(f"   ✓ Filled {missing_names} missing 'Suspect Name' with 'Unknown'")

# Strategy 2: Fill missing ID with generated ID
if 'ID No.' in df_clean.columns:
    missing_ids = df_clean['ID No.'].isnull().sum()
    # Create unique IDs for missing values: "MISSING_0", "MISSING_1", etc.
    df_clean['ID No.'].fillna(
        df_clean['ID No.'].apply(lambda x: f"MISSING_{df_clean.index[df_clean['ID No.'].isnull()].tolist().index(df_clean[df_clean['ID No.'].isnull()].index[0])}" if pd.isnull(x) else x)
    )
    print(f"   ✓ Generated {missing_ids} missing 'ID No.' with unique identifiers")

# Strategy 3: Fill missing gender with most common value (mode)
if 'Suspect Gender' in df_clean.columns:
    missing_gender = df_clean['Suspect Gender'].isnull().sum()
    if missing_gender > 0:
        most_common_gender = df_clean['Suspect Gender'].mode()[0]
        # .mode() finds most frequent value
        # [0] gets the first mode (in case of tie)
        df_clean['Suspect Gender'].fillna(most_common_gender, inplace=True)
        print(f"   ✓ Filled {missing_gender} missing 'Suspect Gender' with '{most_common_gender}' (most common)")

# Strategy 4: Fill categorical columns with 'Unknown'
categorical_cols = ['Suspect Ethnicity', 'Suspect Village', 'Suspect County',
                    'Suspect Area Chief', 'Secondary Reason for Arrest']

for col in categorical_cols:
    if col in df_clean.columns:
        missing = df_clean[col].isnull().sum()
        if missing > 0:
            df_clean[col].fillna('Unknown', inplace=True)
            print(f"   ✓ Filled {missing} missing '{col}' with 'Unknown'")

# Strategy 5: Fill contact numbers with 'Not Provided'
if 'Suspect Contact No.' in df_clean.columns:
    missing_contact = df_clean['Suspect Contact No.'].isnull().sum()
    if missing_contact > 0:
        df_clean['Suspect Contact No.'].fillna('Not Provided', inplace=True)
        print(f"   ✓ Filled {missing_contact} missing 'Suspect Contact No.' with 'Not Provided'")

# Check if cleaning worked
print("\n🔹 After Cleaning:")
remaining_missing = df_clean.isnull().sum().sum()  # Total missing cells across all columns
print(f"   Total missing values remaining: {remaining_missing}")

if remaining_missing == 0:
    print("   ✅ All missing data handled!")
else:
    print(f"   ⚠️ Still {remaining_missing} missing values (might be in Lat/Lon or Date)")
```

```python
# ==================================================
# STEP 4: DATE PROCESSING
# ==================================================

print("\n" + "=" * 70)
print("📅 PROCESSING DATES")
print("=" * 70)

# Convert date column to datetime format
if 'OB No and Date' in df_clean.columns:
    print("\n🔹 Converting 'OB No and Date' to datetime format...")

    # pd.to_datetime() converts text like "2023-05-15" to actual date Python understands
    # errors='coerce' means: if date is invalid, set to NaT (Not a Time) instead of crashing
    df_clean['OB No and Date'] = pd.to_datetime(df_clean['OB No and Date'], errors='coerce')

    # Extract useful parts from the date
    df_clean['Year'] = df_clean['OB No and Date'].dt.year        # Extract year (2023)
    df_clean['Month'] = df_clean['OB No and Date'].dt.month      # Extract month (1-12)
    df_clean['Month_Name'] = df_clean['OB No and Date'].dt.month_name()  # Month as text (January)
    df_clean['Day'] = df_clean['OB No and Date'].dt.day          # Day of month (1-31)
    df_clean['DayOfWeek'] = df_clean['OB No and Date'].dt.day_name()     # Day name (Monday)
    df_clean['Quarter'] = df_clean['OB No and Date'].dt.quarter  # Quarter (Q1, Q2, Q3, Q4)

    print("   ✓ Extracted: Year, Month, Month_Name, Day, DayOfWeek, Quarter")

    # Show date range
    min_date = df_clean['OB No and Date'].min()  # Earliest date
    max_date = df_clean['OB No and Date'].max()  # Latest date
    print(f"\n🔹 Date Range:")
    print(f"   Earliest arrest: {min_date}")
    print(f"   Latest arrest: {max_date}")
    print(f"   Total time span: {(max_date - min_date).days} days")

    # Show distribution by year
    print(f"\n🔹 Arrests by Year:")
    year_counts = df_clean['Year'].value_counts().sort_index()
    # .value_counts() counts occurrences of each unique value
    # .sort_index() sorts by year (2020, 2021, 2022...)
    print(year_counts)
```

```python
# ==================================================
# STEP 8: STATISTICAL HYPOTHESIS TESTING
# ==================================================

from scipy.stats import chi2_contingency, ttest_ind
import pandas as pd

print("\n" + "=" * 70)
print("📊 STATISTICAL ANALYSIS - HYPOTHESIS TESTING")
print("=" * 70)

# ============================================
# TEST 1: Chi-Square Test
# Question: Is there a relationship between Gender and Arrest Reason?
# ============================================

print("\n🔬 TEST 1: Chi-Square Test")
print("=" * 50)
print("Research Question: Do males and females get arrested for different reasons?")
print("\nHypotheses:")
print("  H0 (Null): Gender and arrest reason are INDEPENDENT (no relationship)")
print("  H1 (Alternative): Gender and arrest reason are RELATED")

# Create crosstab (contingency table)
# This shows counts for each combination of gender and reason
crosstab = pd.crosstab(df_clean['Suspect Gender'],
                       df_clean['Primary Reason for Arrest'])

print(f"\n📋 Crosstab (first 5 reasons):")
print(crosstab.iloc[:, :5])  # Show first 5 columns only

# Perform chi-square test
chi2_stat, p_value, dof, expected_freq = chi2_contingency(crosstab)

print(f"\n📊 Chi-Square Test Results:")
print(f"   Chi-square statistic: {chi2_stat:.4f}")
print(f"   P-value: {p_value:.6f}")
print(f"   Degrees of freedom: {dof}")

# Decision
alpha = 0.05  # Significance level
print(f"\n🎯 Decision (alpha = {alpha}):")
if p_value < alpha:
    print(f"   ✅ REJECT H0 (p = {p_value:.6f} < {alpha})")
    print("   Conclusion: Gender and arrest reason ARE related")
    print("   Meaning: Males and females tend to be arrested for different crimes")
else:
    print(f"   ❌ FAIL TO REJECT H0 (p = {p_value:.6f} >= {alpha})")
    print("   Conclusion: No significant relationship found")
    print("   Meaning: Gender doesn't seem to affect type of crime")

# ============================================
# TEST 2: Compare Two Groups
# Question: Do different counties have different arrest rates?
# ============================================

print("\n\n🔬 TEST 2: Comparing Makueni vs Other Counties")
print("=" * 50)

# Create two groups
makueni_arrests = df_clean[df_clean['Suspect County'] == 'Makueni']
other_arrests = df_clean[(df_clean['Suspect County'] != 'Makueni') &
                         (df_clean['Suspect County'] != 'Unknown')]

print(f"Group 1: Makueni County ({len(makueni_arrests)} arrests)")
print(f"Group 2: Other Counties ({len(other_arrests)} arrests)")

# Count top crime for each group
print(f"\nTop crime in Makueni:")
print(makueni_arrests['Primary Reason for Arrest'].value_counts().head(3))

if len(other_arrests) > 0:
    print(f"\nTop crime in Other Counties:")
    print(other_arrests['Primary Reason for Arrest'].value_counts().head(3))

# ============================================
# TEST 3: Summary Statistics by Group
# ============================================

print("\n\n📈 SUMMARY: Arrests by Ethnicity")
print("=" * 50)

# Count arrests by ethnicity (excluding Unknown)
df_known_eth = df_clean[df_clean['Suspect Ethnicity'] != 'Unknown']
ethnicity_summary = df_known_eth['Suspect Ethnicity'].value_counts()

print(ethnicity_summary)

# Calculate percentages
print(f"\n📊 Percentages:")
for ethnicity, count in ethnicity_summary.items():
    pct = (count / len(df_known_eth)) * 100
    print(f"   {ethnicity}: {pct:.1f}%")
```
